## Spliting data into input and output variables

```python
X = dataset[['bedrooms','bathrooms','sqft_living', 'sqft_lot', 'floors','waterfront','view',
             'condition','grade','sqft_above', 'sqft_basement','yr_built','yr_renovated','zipcode',
            'lat','long']]
y = dataset[['price']]
```

```python
y.shape
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

for item in [X_train, y_train, X_test, y_test]:
    print(item.shape)
```

## Fitting Random Forest Regression & Evaluating the model

```python
# Importing models
from sklearn.ensemble import RandomForestRegressor

# Importing metrics module
from sklearn import metrics



# Function for Random Forest regression model for a given n_estimators
def create_Randomforest_regression_model(n_estimators):
    regressor = RandomForestRegressor(n_estimators = n_estimators,random_state=0 )
    regressor.fit(X_train, y_train.values.ravel())
    y_pred = regressor.predict(X_test)
    
 
    # evaluating the model on test dataset
    m = []
    rmse_test = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
    mse_test = metrics.mean_squared_error(y_test, y_pred)
    mae_test = metrics.mean_absolute_error(y_test, y_pred)
    r2_test = metrics.r2_score(y_test, y_pred)
    
    m = [rmse_test,mse_test, mae_test, r2_test]    
    return m
```

```python
rmses = []
n_estimators = np.arange(1, 100)
min_rmse, min_nestimators = 1e10, 1

for n_estimators in n_estimators:
    rmse_test = create_Randomforest_regression_model(n_estimators)[0]
    rmses.append(rmse_test)
    # Cross-validation of degree
    if min_rmse > rmse_test:
        min_rmse = rmse_test
        min_nestimators = n_estimators
        mse_test = create_Randomforest_regression_model(n_estimators)[1]
        mae_test = create_Randomforest_regression_model(n_estimators)[2]
        r2_test = create_Randomforest_regression_model(n_estimators)[3]
# Plot and present results
print('Best n_estimators {} with RMSE {}, MSE {}, MAE {}, R2 {}'.format(min_nestimators,min_rmse,mse_test,mae_test,r2_test))
```
